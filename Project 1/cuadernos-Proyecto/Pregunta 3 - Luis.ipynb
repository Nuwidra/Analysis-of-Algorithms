# Plan de investigación:

## Problema:

¿Cómo se compara el algoritmo *quicksort* con el algoritmo *mergesort* en términos de rendimiento temporal?

## Hipótesis:

Se comparan mediante el analisis en sus algoritmos y se estable que el algoritmo mas eficiente es el de `QuickSort` ya que la cantidad de pasos a ejecutar en su algoritmo es mucho menor al  `Merge Sort` siendo este el que ejecuta una mayor cantidad de pasos la cual esta relacionada a la entrada que tengan como paramentro.

## Metodología:

Se realizarán las pruebas en tres diferentes tipos de conjuntos y cada uno con dos tamaños distintos en total se darían 6 corridas distintas:
1. Mejor caso para **QuickSort** con pocos elementos(100) y con muchos elementos(1000).
2. Peor caso para **QuickSort** con pocos elementos(100) y con muchos elementos(1000).
3. Caso promedio para **QuickSort** con pocos elementos(100) y con muchos elementos(1000).


Cada prueba será ejecutando los algoritmos *quicksort* y *mergesort* con cada conjunto, utilizando `big_o` para mostrar los resultados del tiempo de ejecución y posteriormente el analisis según la información mediante la observación de gráficas y tablas con los tiempos de ejecución y la variación entre ellos, para así dictaminar claramente cual es el mecanismo mas eficiente.

Otro parametro importante es la aderencia al criterio supuesto de comportamiento viendose en la aproximación asintótica la cual entre mas cercana a `1` será mucho más precisa la aproximación

```python
from big_o import big_o
from big_o import complexities as cls

import random
import numpy as np
import pandas as pd
```

```python
def aprox_asintotica(datos):
    ys = datos['times']
    for k, v in datos.items():
        if isinstance(k, cls.ComplexityClass):
            residual = v
            r2 = 1 - residual / (ys.size * ys.var())
            print(k, f' (r={residual}, r^2={r2})')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

def plot(etiqueta_xs, xs, etiqueta_ys, ys):
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    ax.bar(xs, ys)
    plt.ylabel(etiqueta_ys)
    plt.xlabel(etiqueta_xs)
    plt.show()

def plot_hist(etiqueta_xs, etiqueta_ys, ys):
    fig, ax = plt.subplots()
    ax.hist(ys, bins=10)
    plt.xlabel(etiqueta_xs)
    plt.ylabel(etiqueta_ys)
    plt.show()
```

### 1. Mejor caso para Quick Sort y corrida en Merge Sort(no tiene casos).

#### 1.a

**Quick Sort**

```python
clase, BST_QS = big_o(lambda A: quicksort_mejor(A), 
                     ordenada_ascendente,
                     min_n=10, max_n=500,
                     n_measures=490, n_repeats=3,
                     verbose=False, return_raw_data=True,
                     classes=[cls.Linearithmic])
```

```python
bst_qs_xs = BST_QS['measures']
bst_qs_ys = BST_QS['times']
bst_qs_aprox_ys = clase.compute(bst_qs_xs)

plt.plot(bst_qs_xs, bst_qs_ys, label = 'Quick Sort')
plt.plot(bst_qs_xs, bst_qs_aprox_ys, label = 'Aprox')
plt.ylabel('Tiempo (s)')
plt.xlabel('n (Tamaño del arreglo)')
plt.legend(loc = 2)
plt.show()
```

```python
aprox_asintotica(BST_QS)
```

**Merge Sort:**

```python
clase, BST_MS = big_o(lambda A: merge_sort(A), 
                     ordenada_ascendente,
                     min_n=10, max_n=500,
                     n_measures=490, n_repeats=3,
                     verbose=False, return_raw_data=True,
                     classes=[cls.Linearithmic])
```

```python
bst_ms_xs = BST_MS['measures']
bst_ms_ys = BST_MS['times']
bst_ms_aprox_ys = clase.compute(bst_ms_xs)

plt.plot(bst_ms_xs, bst_ms_ys,label = 'Merge sort')
plt.plot(bst_ms_xs, bst_ms_aprox_ys, label = 'Aprox')
plt.ylabel('Tiempo (s)')
plt.xlabel('n (Tamaño del arreglo)')
plt.legend(loc = 2)
plt.show()
```

```python
aprox_asintotica(BST_MS)
```

**Comparativa:**

```python
plt.plot(bst_qs_xs, bst_qs_ys, label = 'Quick Sort')
plt.plot(bst_qs_xs, bst_qs_aprox_ys, label = 'Quick Sort aprox')
plt.plot(bst_ms_xs, bst_ms_ys, label = 'Merge Sort')
plt.plot(bst_ms_xs, bst_ms_aprox_ys, label = 'Merge Sort aprox')
plt.ylabel('Tiempo (s)')
plt.xlabel('n (Tamaño del arreglo)')
plt.legend(loc = 2)
plt.show()
```

```python
bst = pd.DataFrame({
    'n': bst_qs_xs,
    'Quick Sort': bst_qs_ys,
    'Quick Sort aprox': bst_qs_aprox_ys,
    'Merge Sort': bst_ms_ys,
    'Merge Sort aprox': bst_ms_aprox_ys,
})

bst['prop'] = bst['Quick Sort aprox'] / bst['Merge Sort aprox']
```

```python
display(bst)
```
